```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/kddcup.csv")
df
```

```python
df.describe()
```

There is an missing value respectively in `src_bytes` and `dst_bytes`, fill them with the mean value of each columns:

```python
df['src_bytes'] = df['src_bytes'].fillna(3.025610e+03)
df['dst_bytes'] = df['dst_bytes'].fillna(8.685324e+02)
df.isnull().sum()
```

```python
df['type'].unique()
```

```python
df
```

```python
cat_features = []
for col in df:
    if df[col].dtypes == 'object':
        cat_features.append(col)
cat_features
```

```python
for cat_feature in cat_features:
    for feature in df[cat_feature].unique():
        df[cat_feature+'_is_'+str(feature)] = [0]*df.shape[0]
```

```python
df
```

```python
##  Please DO NOT try to run this chunk!! It will take you hours!!

for i in range(df.shape[0]):
    for cat_feature in cat_features:
        for feature in df[cat_feature].unique():
            if df.iloc[i][cat_feature] == feature:
                df.at[i, cat_feature+'_is_'+str(feature)] = 1
    print(i) # just to see how long it will take
```
